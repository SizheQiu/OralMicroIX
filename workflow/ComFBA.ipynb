# A bin.28 Haemophilus

```python
def v_group(v):
    if 'bin' in v:
        return 'Bacteria'
    else:
        return 'Metabolite'
    
def build_G(table, name, scale):
    G=nx.Graph(name =name); edge_list=[];
    vs = list( np.unique( list(table.receiver)+list(table.donor) ) ) + list(table.compound)
    G.add_nodes_from([(v,{'group':v_group(v)}) for v in vs])
    for i in range(len(table.index)):
        r, d, met, _, _, _, weight = table.iloc[i]
        edge_list.append( (r,met,{'weight':abs(weight)*scale,'edgecolor':'tab:red'}) )
        edge_list.append( (d,met,{'weight':abs(weight)*scale,'edgecolor':'tab:blue'}) )
    G.add_edges_from(edge_list)
    return G
```

```python
rm_compounds = ['fe3','fe2','o2','h','h2o','fe3pyovd_kt']
A_crossfeed = pd.read_csv('../data/metagenome/A_smetana.tsv',sep='\t')
comps= list(A_crossfeed['compound'])
A_crossfeed['compound'] = [x.replace('M_','').replace('_e','') for x in comps]
A_crossfeed = A_crossfeed[['receiver', 'donor', 'compound', 'scs', 'mus','mps', 'smetana']]
```

```python
A_bin28 = A_crossfeed[(A_crossfeed.receiver=='A_bin.28')|(A_crossfeed.donor=='A_bin.28')]
A_bin28 = A_bin28[~A_bin28.compound.isin(rm_compounds)].reset_index().drop(['index'],axis=1)
```

```python
G_A28 = nx.Graph(name = 'binA28')
vs = list( np.unique( list(A_bin28.receiver)+list(A_bin28.donor) ) ) + list(A_bin28.compound)
G_A28.add_nodes_from([(v,{'group':v_group(v)}) for v in vs])
edge_list=[]
for i in range(len(A_bin28.index)):
    r, d, met, _, _, _, weight = A_bin28.iloc[i]
    edge_list.append( (r,met,{'weight':abs(weight)*50,'edgecolor':'tab:red'}) )
    edge_list.append( (d,met,{'weight':abs(weight)*50,'edgecolor':'tab:blue'}) )
G_A28.add_edges_from(edge_list)
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,6))
nodecolors = []
for v in G_A28.nodes:
    if 'bin' in v:
        nodecolors.append('tab:orange')
    else:
        nodecolors.append('plum')      
edgecolors = [G_A28[u][v]['edgecolor'] for u,v in G_A28.edges]
weights = [G_A28[u][v]['weight'] for u,v in G_A28.edges]
nx.draw_circular( G_A28, edge_color=edgecolors, node_color=nodecolors, edgecolors='black', \
                 width=weights, with_labels = True,font_weight='bold',alpha=0.7,ax=ax)
# plt.savefig('../../../../Oral_paper/Figures_2025/comfba_A28.svg',format='svg',dpi=150)
```

# Species coupling

```python
A_crossfeed = pd.read_csv('../data/metagenome/A_smetana.tsv',sep='\t')
A_scs = A_crossfeed[['receiver', 'donor', 'scs']].drop_duplicates().reset_index().drop(['index'],axis=1)
A_scs = A_scs[A_scs.scs>0.1].reset_index().drop(['index'],axis=1)
```

```python
G_A = nx.DiGraph(name = 'A')
vs = list( np.unique( list(A_scs.receiver)+list(A_scs.donor) ) )
G_A.add_nodes_from([(v) for v in vs])
edge_list=[]
for i in range(len(A_scs.index)):
    r,d,weight=A_scs.iloc[i]
    edge_list.append( (d,r,{'weight':abs(weight)*5} ) )
G_A.add_edges_from(edge_list)
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,10))
weights = [G_A[u][v]['weight'] for u,v in G_A.edges]

coms_A = nx.community.greedy_modularity_communities(G_A)
supergraph = nx.cycle_graph(len(coms_A))
superpos = nx.spring_layout(G_A, scale=5,seed=0)
centers = list(superpos.values())
pos = {}
for center, com in zip(centers, coms_A):
    pos.update(nx.spring_layout(nx.subgraph(G_A, com), center=center, seed=0))
for nodes in coms_A:
    print(list(nodes))
    nx.draw_networkx_nodes(G_A, pos=pos, nodelist=nodes, node_color='tab:purple',alpha=0.7,
                           edgecolors='black',node_size=100,ax=ax)
nx.draw_networkx_edges(G_A, pos=pos,edge_color='tab:gray', width=weights, alpha=0.8,
                       connectionstyle="arc3,rad=0.1",ax=ax)
nx.draw_networkx_labels(G_A, pos=pos,font_weight='bold',font_size=8 );
# plt.savefig('../../../../Oral_paper/Figures_2025/A_scs.svg',format='svg',dpi=150)
```
