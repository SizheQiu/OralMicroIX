# UMAP of pathways

```python
path_abund = pd.read_csv('../data/metagenome/pathabundance.csv')
path_abund = path_abund.set_index(['Pathway'])
path_abund = path_abund.drop(['UNINTEGRATED','UNMAPPED'],axis=0)
```

```python
path_abund_T = path_abund.T
X = path_abund_T.iloc[:,:].values
X = StandardScaler().fit_transform(X)
reducer_path = umap.UMAP(random_state=1)
embedding = reducer_path.fit_transform(X)
umap_pd = pd.DataFrame(embedding, columns =['UMAP-1','UMAP-2'])
umap_pd['Group'] = ['A (20~30yo)' for i in range(10)] + ['B (55~80yo)' for i in range(10)]
```

```python
fig, ax = plt.subplots(1,1,figsize=(6,5.5))
sns.scatterplot(x='UMAP-1', y='UMAP-2',hue="Group",
            linewidth=1,data=umap_pd, s=300, alpha=0.7, edgecolor='black', ax=ax)
ax.set_xlabel('UMAP-1'); ax.set_ylabel('UMAP-2');
ax.set_title('Metabolic pathway abundance',fontsize=12)
# plt.savefig('../../../../Oral_paper/Figures_2025/umap_pathway.svg',format='svg',dpi=150)
```

```python
gene_deg = pd.read_csv('../data/metagenome/gene_deg.csv')
gene_deg = gene_deg.rename(columns={'Sample':'Gene'})
gene_deg['Gene'] = [x.split('_')[1].strip() for x in list(gene_deg['Gene'])]
drop_inds=[]
for i in range(len(gene_deg.index)):
    if 'A0A' in list(gene_deg['Gene'])[i]:
        drop_inds.append(i)
gene_deg = gene_deg.drop(drop_inds).reset_index().drop(['index'],axis=1)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
volcano( gene_deg, 'log2FoldChange', 'padj', 1.0, 0.05, 150, ax=ax)
ax.set_xlabel('Log2 Fold Change',fontsize=14)
ax.set_ylabel('- Log10(P-value)',fontsize=14)
# ax.set_ylim(0.8,3.2); ax.set_xlim(-1.75,2.8);
annot_volcano( gene_deg,'Gene' ,'log2FoldChange', 'padj', 1.0, 0.005, 10, ax=ax );
# plt.savefig('../../../../Oral_paper/Figures_2025/daa_genes.svg',format='svg',dpi=150)
```

```python
lfc_col='log2FoldChange'; x_cut1=1.0;x_cut2=-1.0;pv_col='padj';pv_cutoff=0.05;
table =gene_deg
sig_table = table[ ( (table[lfc_col]>x_cut1) | (table[lfc_col]<x_cut2) ) & (table[pv_col]<pv_cutoff)  ]
sig_table = sig_table.sort_values(['log2FoldChange'],ascending=False)
sig_table = sig_table.reset_index().drop(['index'],axis=1)
```

```python
def get_go_annotations(uniprot_id):
    url = f"https://rest.uniprot.org/uniprotkb/{uniprot_id}.json"

    try:
        # Make a GET request to the UniProt API
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()
        go_terms=[]
        if 'uniProtKBCrossReferences' not in data.keys():
            return 'None'
        go_annots = [k for k in data['uniProtKBCrossReferences'] if k['database']=='GO']
        if len(go_annots)<1:
            return 'None'
        else:
            for item in go_annots:
                temp = item['properties'][0]['value']
                if 'P:' in temp:
                    go_terms.append(temp)
        return go_terms

    except requests.exceptions.RequestException as e:
        print(f"Error fetching data for UniProt ID {uniprot_id}: {e}")
        return 'None'
```

```python
go_dict = {}
for pid in list(sig_table['Gene']):
    go_terms = get_go_annotations(pid)
    if go_terms != 'None' and len(go_terms)>=1:
        go_dict[pid]=';'.join(go_terms)
```

```python
go_count = {}
for k in go_dict:
    go_terms = go_dict[k].split(';')
    for term in go_terms:
        if term in go_count.keys():
            go_count[term] += 1
        else:
            go_count[term] = 1
```

```python
go_stats = pd.DataFrame(go_count.items(),columns=['GO','Count'])
go_stats = go_stats.sort_values(by=['Count'],ascending=False).reset_index().drop(['index'],axis=1)
# go_stats.to_csv('../data/metagenome/gene_deg_go.csv',index=None)
```

```python
go_stats_top=go_stats.iloc[:10]
fig, ax = plt.subplots(1,1,figsize=(7,6))
ax.barh(list(go_stats_top['GO']), list(go_stats_top['Count']),color='tab:red')
ax.set_xlabel('Count')
ax.set_title('GO enrichment of up-regulated genes')
ax.text(x=6,y=8,s='B vs A',fontsize=14)
# plt.savefig('../../../../Oral_paper/Figures_2025/daa_gene_go.svg',format='svg',dpi=150)
```
